```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10 # degree of freedom

# Create x-axis data
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 1000)

# Create y-axis data
y = t.pdf(x, df)

# Plot the distribution
plt.plot(x, y, 'r-', lw=2, alpha=0.6, label='t pdf')

plt.title('Student t-distribution with 10 degrees of freedom')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.legend(loc='best', frameon=False)
plt.show()
```
